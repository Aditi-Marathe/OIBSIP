```python
df.rename(columns={
    'Region': 'State',
    'Estimated U': 'Unemployment_Rate',
    'Estimated E': 'Estimated_Employed',
    'Estimated LF': 'Estimated_Labour_Force'
}, inplace=True)
```

```python
# Clean spaces from column
df.columns = df.columns.str.strip()


df.columns
```

```python
#Convert Date to datetime

# 1. Convert Date column to string and remove spaces
df['Date'] = df['Date'].astype(str).str.strip()

# 2. Convert Date column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
```

```python
df['Date'].head(10)
```

```python
# Rename the columns names

df.rename(columns={
    'Estimated U': 'Unemployment Rate',
    'Estimated E': 'Employed',
    'Estimated LF': 'Labour Force',
    'Region.1': 'Zone'
}, inplace=True)
```

```python
df.rename(columns={
    'Estimated Unemployment Rate (%)': 'Unemployment_Rate',
    'Estimated Employed': 'Employed',
    'Estimated Labour Participation Rate (%)': 'Labour_Participation_Rate'
}, inplace=True)
```

### **Step 4: Visualize the dataset**

```python
df.columns
```

```python
# Line plot of overall unemployment rate over time

plt.figure(figsize=(12,6))
sns.lineplot(data=df, x='Date', y='Unemployment_Rate')
plt.title('Overall Estimated Unemployment Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.grid(True)
plt.show()
```

```python
# Bar plot of average unemployment rate by Region

region_avg = df.groupby('Zone')['Unemployment_Rate'].mean().sort_values(ascending=False)
plt.figure(figsize=(10,6))
sns.barplot(x=region_avg.values, y=region_avg.index, palette='viridis')
plt.title('Average Estimated Unemployment Rate by Region')
plt.xlabel('Average Unemployment Rate (%)')
plt.ylabel('Region')
plt.show()
```

```python
# Monthly average unemployment rate trend

df['Month'] = df['Date'].dt.to_period('M')
monthly_avg = df.groupby('Month')['Unemployment_Rate'].mean()

plt.figure(figsize=(10,5))
monthly_avg.plot(marker='o', color='purple')
plt.title('Monthly Average Estimated Unemployment Rate')
plt.xlabel('Month')
plt.ylabel('Unemployment Rate (%)')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x='Zone', y='Unemployment_Rate')
plt.title('Distribution of Unemployment Rate by Region')
plt.xticks(rotation=45)
plt.show()
```

### **Step 5: Analyzing the sharp increase in unemployment**

```python
# Overall Unemployment Rate Trend

plt.figure(figsize=(12,6))
sns.lineplot(data=df, x='Date', y='Unemployment_Rate', marker='o', color='red')
plt.axvspan(pd.Timestamp('2020-03-01'), pd.Timestamp('2020-07-31'), color='gray', alpha=0.3, label='COVID-19 Lockdown Period')
plt.title('Unemployment Rate Trend Over Time (Highlighting COVID-19)')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.legend()
plt.grid(True)
plt.show()
```

There is a sharp spike in unemployment starting around March 2020, coinciding with the national lockdown and COVID-19 onset, followed by a gradual decline after July 2020.

```python
# Region-wise Unemployment Rate Trend During COVID

plt.figure(figsize=(14,7))
covid_period = (df['Date'] >= '2020-03-01') & (df['Date'] <= '2020-07-31')
sns.lineplot(data=df[covid_period], x='Date', y='Unemployment_Rate', hue='Zone')
plt.title('Region-wise Unemployment Rate During COVID-19 Lockdown Period')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.legend(bbox_to_anchor=(1.05,1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

During the COVID lockdown, some regions (e.g., Assam) faced more severe unemployment spikes than others (e.g., Andhra Pradesh), indicating regional economic resilience varied.

```python
#  Monthly Average Unemployment Rate

df['Month'] = df['Date'].dt.to_period('M')
monthly_avg = df.groupby('Month')['Unemployment_Rate'].mean()

plt.figure(figsize=(10,5))
monthly_avg.plot(marker='o', color='blue')
plt.axvline(pd.Period('2020-03'), color='red', linestyle='--', label='Start of COVID Lockdown')
plt.axvline(pd.Period('2020-08'), color='green', linestyle='--', label='End of Lockdown')
plt.title('Monthly Average Unemployment Rate (Pre, During, Post COVID)')
plt.xlabel('Month')
plt.ylabel('Unemployment Rate (%)')
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

The unemployment rate was stable before March 2020, sharply increased during lockdown, and started declining post-August 2020 as restrictions eased.

```python
# ncrease in Average Unemployment Rate Before vs During COVID (By Region)

pre_covid = df[df['Date'] < '2020-03-01'].groupby('Zone')['Unemployment_Rate'].mean()
during_covid = df[(df['Date'] >= '2020-03-01') & (df['Date'] <= '2020-07-31')].groupby('Zone')['Unemployment_Rate'].mean()
increase = during_covid - pre_covid

increase = increase.sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=increase.values, y=increase.index, palette='magma')
plt.title('Increase in Average Unemployment Rate During COVID vs Pre-COVID by Region')
plt.xlabel('Increase in Unemployment Rate (%)')
plt.ylabel('Region')
plt.show()
```

This graph clearly shows South regions experienced the biggest jump in unemployment during COVID compared to before. Some regions had increases over 20%, highlighting the pandemic's uneven impact.

```python
# Distribution of Unemployment Rates Before and During COVID
df['Period'] = df['Date'].apply(lambda x: 'Pre-COVID' if x < pd.Timestamp('2020-03-01') else ('During COVID' if x <= pd.Timestamp('2020-07-31') else 'Post-COVID'))

plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Period', y='Unemployment_Rate', palette='Set1')
plt.title('Unemployment Rate Distribution Before, During and After COVID')
plt.xlabel('Period')
plt.ylabel('Unemployment Rate (%)')
plt.show()
```

The boxplot reveals how the median and spread of unemployment rate drastically increased during COVID compared to before and stabilized somewhat post-COVID

```python
#Correlation Between Unemployment and Other Indicators During COVID


covid_df = df[(df['Date'] >= '2020-03-01') & (df['Date'] <= '2020-07-31')]

corr = covid_df[['Unemployment_Rate', 'Employed', 'Labour_Participation_Rate']].corr()

print(corr)

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix during COVID-19 Period')
plt.show()
```

The heatmap shows a strong negative correlation between unemployment and employment, as expected, reinforcing that higher unemployment meant fewer employed persons during the pandemic.

```python
# Unemployment Rate During Peak COVID Month

peak_covid_month = covid_df[covid_df['Date'].dt.month == 4]  # April 2020 peak

plt.figure(figsize=(10,7))
sns.scatterplot(
    data=peak_covid_month,
    x='longitude',
    y='latitude',
    size='Unemployment_Rate',
    hue='Unemployment_Rate',
    palette='Reds',
    alpha=0.7,
    sizes=(50, 300)
)
plt.title('Geographical Spread of Unemployment Rate (April 2020 - COVID Peak)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()

```

This spatial plot highlights regions with highest unemployment rates geographically during the peak COVID month (April 2020), showing which states or districts were most affected

The analysis of the unemployment rate over time clearly highlights the significant impact of the COVID-19 pandemic on the labor market. The sharp rise in unemployment during the early months of the pandemic, particularly in April 2020, reflects the widespread economic disruptions caused by lockdowns and business closures.
